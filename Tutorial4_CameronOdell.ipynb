e. Is there a relationship between **house price** and number of **bedrooms** in any plot?

```python
# To-do: Try any plot with "HousePrices" dataset and describe your findings 
# (Is there a relationship between house price and number of bedrooms in any plot?)

# Create a boxplot of house prices by number of bedrooms
sns.boxplot(x=houseprices_df['bedrooms'], y=houseprices_df['price'])

# Show the plot
plt.show()

# Correlation coefficient of price & bedrooms
bedroom_price_corr = houseprices_df['price'].corr(houseprices_df['bedrooms'])
bedroom_price_corr_rounded = round(bedroom_price_corr, 4)
print(f'The correlation coefficient between price and bedrooms without assessing other variables or removing outliers is {bedroom_price_corr_rounded}')
```

**Findings for Task I.e.:**

I decided a box & whisker plot would be the best way to represent the data type since there are only a few categories and its easier to visualize what the range of prices look like with quartiles and median values. Based on this plot, it appears there is a positive linear relationship as bedrooms increase with the price also increasing. This holds true except for 6 bedroom homes where the median price is equivalent 3 bedroom home.

Using a correlation coefficient the positive relationship r is approximately 0.3664. This is a weak to moderate relationship between number of bedrooms and price increase, but definitely not insignificant. While the number of bedrooms may increase home price, there are likely other factors that are equally significant or more significant in driving the value of price in a home.

f. Is there a relationship between **house price** and **lotsize** in any plot?

```python
# To-do: Try any plot with "HousePrices" dataset and describe your findings 
# (Is there a relationship between house price and lotsize in any plot?)

# Defining scatter plot with regression line using seaborn regression plot
sns.regplot(x=houseprices_df['lotsize'], y=houseprices_df['price'], 
            line_kws={"color":"black","alpha":0.7,"lw":1},
            scatter_kws={"s": 5}, ci=None)

# Calculate expected y-values for each x-value
x = houseprices_df['lotsize']
y = houseprices_df['price']
slope, intercept = np.polyfit(x, y, 1)
y_expected = slope * x + intercept

# Assign colors based above & below expected price on the regression line
colors = np.where(y > y_expected, 'red', 'blue')
sns.scatterplot(x=x, y=y, hue=colors, palette=['blue', 'red'], s=5)
plt.legend(handles=[
    Line2D([0], [0], marker='o', color='w', label='Above Expected Price', markerfacecolor='red', markersize=5),
    Line2D([0], [0], marker='o', color='w', label='Below Expected Price', markerfacecolor='blue', markersize=5)
])
plt.show()

# Correlation Coefficient of price and lotsize
lotsize_price_corr = houseprices_df['price'].corr(houseprices_df['lotsize'])
lotsize_price_corr_rounded = round(lotsize_price_corr, 4)
print(f'The correlation coefficient between price and lotsize without assessing other variables or removing outliers is {lotsize_price_corr_rounded}')
```

**Findings for Task I.f.:**

There appears to be a moderate postive correlation between lotsize and price of a home. The correlation coeffient in r is approximately +0.5358. There is a stronger association between lotsize and price than number of bedrooms and price. However, there are likely other variables that hold equal weight or potentially larger weight in driving the price of a home than lotsize. This linear regression is represented by the above scatterplot. Note that many of the data points don't fit close to the regression line, but it does show a significant positive correlation. There also appear to be likely outliers that could be removed from the data set to create a better fitting model.
